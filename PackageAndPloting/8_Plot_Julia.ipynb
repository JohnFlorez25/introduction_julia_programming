# Graficando en Julia (Plots)
Existen varias formas de realizar gráficas en Julia (incluyendo PyPlot)

Vamos a ver como es el uso de `Plots.jl`

```julia
using Pkg
```

```julia
Pkg.add("Plots")
```

```julia
using Plots
```

Una de las ventajas al usar `Plots.jl` es que te permite cambiar sin problemas los backends. Vamos a probar `gr()` y `plotlyjs()` backends.

Primero, vamos a generar algunos datos aleatorios para graficarlos

```julia
x = -3:0.1:3

f(x) = x^2

y = f.(x)
```

### Ahora vamos a cargar el backend GR

```julia
gr()
```

```julia
plot(x, y, label="line")
scatter!(x, y, label="points")
```

El `!` al final de la función `scatter!` lo  convierte en una función que muta, indicando  que los puntos dispersos van a ser agregador en una gráfica pre existente.

En contraste, observa que pasa cuando tu reemplazas `scatter!` por `scatter` 

```julia
plot(x, y, label="line")
scatter(x, y, label="points")
```

###  Graficando cada vez más elegante
La sintaxis para agregar títulos y etiquetas para nuestras gráficas es bastante sencilla.

En estos momentos, en nombre de la investigación científica, vamos a examinar la relación que éxiste entre la temperatura global y el número de piratas entre aproximadamente 1860 y 2000

```julia
globalTemperatures = [14.4, 14.5, 14.8, 15.2, 15.5, 15.8]
numberOfPirates = [45000, 20000, 15000, 5000, 400, 17 ]
```

```julia
# Primera gráfica de los datos
plot(numberOfPirates, globalTemperatures, legend=false)
scatter!(numberOfPirates, globalTemperatures, legend=false)

# Esta funcion genera el eje inverso de x y tu puedes ver como la temperatura cambia
xflip!()

# Agregando etiquetas (labels) y titles (títulos)
xlabel!("Number of Pirates [Aproximmate]")
ylabel!("Global Temperature (C)")
title!("Influence of pirate population on global warming")
```

Para crear gráficas con subgráficas, solo debemos nombrar cada una de las sub-gráficas y luego envolver esos nombres, junto con una especificación de diseño, en otro `plot`.

```julia
p1 = plot(x, x)
p2 = plot(x, x.^2)
p3 = plot(x, x.^3)
p4 = plot(x, x.^4)
plot(p1,p2,p3,p4,layout=(2,2),legend=false)
```
